```python
uniprot_ref = 'SH2_IPR000980_uniprot_reference.csv'
df_ref = pd.read_csv(uniprot_ref)
identical_arch_groups = []
for i in cluster_dict:
    arch_arr = []
    for j in cluster_dict[i][1]:
        gene = (j.split('|')[0])
        arch = df_ref.loc[df_ref['Gene'] == gene, ['Interpro Domain Architecture']].values[0][0]
        arch_arr.append(arch)
    cluster_dict[i][1].append(set(arch_arr))
    if len(set(arch_arr)) == 1:
        identical_arch_groups.append(i)
```

```python
contacting_domains = []
for line in open('SH2_IPR000980_interDomain_Exp.feature'):
    line = line.strip('\n')
    line = line.split('\t')
    if len(line) > 2:
        gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
        gene_dom_pair = line[0] + ':' + gene
        if gene_dom_pair not in contacting_domains:
            contacting_domains.append(gene_dom_pair)

contacting_domains_af = []
for line in open('SH2_IPR000980_interDomain_Alphafold.feature'):
    line = line.strip('\n')
    line = line.split('\t')
    if len(line) > 2:
        gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
        gene_dom_pair = line[0] + ':' + gene
        if gene_dom_pair not in contacting_domains_af:
            contacting_domains_af.append(gene_dom_pair)
```

```python
feafile_af = 'SH2_IPR000980_interDomain_Alphafold.feature'
feafile_exp = 'SH2_IPR000980_interDomain_Exp.feature'
dict_known_features_af = {}

genes_with_features_af = []
for line in open(feafile_af):
    line = line.strip('\n')
    line = line.split('\t')
    if len(line) > 2:
        get_gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
        if get_gene not in genes_with_features_af:
            genes_with_features_af.append(get_gene)

genes_with_features_exp = []
for line in open(feafile_exp):
    line = line.strip('\n')
    line = line.split('\t')
    if len(line) > 2:
        get_gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
        if get_gene not in genes_with_features_exp:
            genes_with_features_exp.append(get_gene)
```

```python
dict_known_features_af = {}
for known_pair in contacting_domains_af:
    feature_position_aln = []
    for line in open(feafile_af):
        line = line.strip('\n')
        line = line.split('\t')
        if len(line) > 2:
            get_gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
            gene_dom_pair = line[0] + ':' + get_gene

            pair, gene = known_pair.split(':')

            if known_pair == gene_dom_pair:
                feature_position_aln.append(line[3])
    #             print(gene, pair, line[3])
    feature_position_aln = [int(i) for i in feature_position_aln]
    dict_known_features_af[known_pair] = feature_position_aln

dict_known_features_exp = {}
for known_pair in contacting_domains:
    feature_position_aln = []
    for line in open(feafile_exp):
        line = line.strip('\n')
        line = line.split('\t')
        if len(line) > 2:
            get_gene = line[1].split('|')[1] + '|' + line[1].split('|')[2]
            gene_dom_pair = line[0] + ':' + get_gene

            pair, gene = known_pair.split(':')

            if known_pair == gene_dom_pair:
                feature_position_aln.append(line[3])
    #             print(gene, pair, line[3])
    feature_position_aln = [int(i) for i in feature_position_aln]
    dict_known_features_exp[known_pair] = feature_position_aln
```

```python
for i in cluster_dict:
    fea_arr = []
    for j in cluster_dict[i][1][:-1]:
        if j in genes_with_features_exp:
            fea_arr.append('E')
        elif j in genes_with_features_af and j not in genes_with_features_exp:
            fea_arr.append('A')
        elif j not in genes_with_features_af and j not in genes_with_features_exp:
            fea_arr.append(0)
    (cluster_dict[i][1].append(fea_arr))
```
